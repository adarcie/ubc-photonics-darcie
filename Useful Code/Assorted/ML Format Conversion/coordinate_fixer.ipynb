```python
#convert coordinates to Maple Leaf format
# - input coords in format input_file.txt
# - outputs as input_file_fixed.txt
# - add_ind is boolean -> set true if you want to add incremented numbers to the end
def coords_fixer(input_file,add_ind=False):
    filename=input_file
    f = open(filename+'.txt')
    line = f.readline()

    f2=open(filename+'_fixed.txt',"w+")

    k=0
    while line:
        if line.split(' ')[0]!="Number" and line!='\n':
            linelist=line.split(', ')
            newstring=''

            if k==3:
                print(line)
                f2.write(line)

            if k>3:

                for i in range(0,5):
                    newstring=newstring+linelist[i]+', '
                for i in range(5,len(linelist)):
                    if i>5:
                        newstring=newstring+'_'+linelist[i]
                    else:
                        newstring=newstring+linelist[i]

                if add_ind:
                    print(newstring.split('\n')[0]+"_"+str(k-3)+'\n')
                    f2.write(newstring.split('\n')[0]+"_"+str(k-3)+'\n')
                else:
                    print(newstring)
                    f2.write(newstring)

        line = f.readline()
        k=k+1
    f.close()
    f2.close()
    
```

```python
coords_fixer('fracture_v5_arrayed_coords',True)
```
